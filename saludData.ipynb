```python
# coding: utf-8

import pandas as pd
from urllib.parse import quote

# Pull tables from Wikipedia
base  = "https://es.wikipedia.org/wiki/" 
path  = "Anexo:Ministros_de_Salud_del_Perú"
url   = base + quote(path)          # Per%C3%BA

# Add a User-Agent header to mimic a browser and avoid 403 Forbidden error
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
saludTables = pd.read_html(url, flavor="bs4",
                      attrs={"class": "wikitable sortable"}, storage_options={"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
```

```python
saluda=saludTables[0]
saluda.columns
```

```python
saluda.columns=[b for a,b in saluda.columns]
```

```python
salud=saluda.copy()
salud.drop(columns=["Imagen"],inplace=True)
```

# date text cleaning

```python
salud=salud[~salud["Inicio"].isnull()]
```

```python
salud.reset_index(drop=True,inplace=True)
salud.info()
```

```python
salud[salud.Inicio.str.contains(r"del|\[",regex=True)]
```

```python
salud['Inicio']=salud['Inicio'].str.replace(r'\[.*\]', '', regex=True)
```

```python
salud[salud.Final.str.contains(r"del|\[",regex=True)]
```

```python

salud['Final']=salud['Final'].str.replace(r'\[.*\]', '', regex=True)
```

```python
import numpy as np
from datetime import datetime, timedelta
import locale

# 1. Configurar locale (con fallback por si estás en Windows)
try:
    locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")
except locale.Error:
    locale.setlocale(locale.LC_TIME, "Spanish_Spain.1252") # Windows

# Tu función lambda (agregamos validación por si hay NaN)
str_to_date = lambda x: datetime.strptime(x, "%d de %B de %Y") if pd.notna(x) else pd.NaT

salud[["Inicio", "Final"]] = salud[["Inicio", "Final"]].apply(lambda x: x.str.strip())
```

# Date format

```python
# 2. Convertir a datetime
salud['fecha_start'] = salud['Inicio'].apply(str_to_date)
salud['fecha_end'] = salud['Final'].apply(str_to_date)



```
