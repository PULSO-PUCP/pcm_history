```python
# coding: utf-8

import pandas as pd
from urllib.parse import quote

# Pull tables from Wikipedia
base  = "https://es.wikipedia.org/wiki/" 
path  = "Anexo:Ministros_del_Interior_del_Perú"
url   = base + quote(path)          # Per%C3%BA

# Add a User-Agent header to mimic a browser and avoid 403 Forbidden error
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
intTables = pd.read_html(url, flavor="bs4",
                      attrs={"class": "wikitable sortable"}, storage_options={"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
```

# Date format

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import locale

# 1. Configurar locale (con fallback por si estás en Windows)
try:
    locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")
except locale.Error:
    locale.setlocale(locale.LC_TIME, "Spanish_Spain.1252") # Windows

# Tu función lambda (agregamos validación por si hay NaN)
str_to_date = lambda x: datetime.strptime(x, "%d de %B de %Y") if pd.notna(x) else pd.NaT

# --- Supongamos que tu columna de fechas se llama 'fecha_inicio' ---

# 2. Convertir a datetime
inta['fecha_start'] = inta['Fecha de inicio'].apply(str_to_date)

# 3. Desplazar (-1), restar 1 día y volver a formato texto español
inta['fecha_end'] = (inta['fecha_start'].shift(-1) - timedelta(days=1))



inta
```
