```python

from pandas.tseries.offsets import QuarterEnd

def split_spell_row(row):
    intervals = []
    current = row['start'].to_period('Q').to_timestamp(how='start')
    while current <= row['end']:
        stop = min((current + QuarterEnd(0)).normalize(), row['end'])
        intervals.append({
            'id': row['id'],
            'start': current,
            'stop': stop,
            'event': int((stop == row['end']) and row['status'] == 1),
            'regimeFujimori': row['regimeFujimori'],
            'technocrat_trait_ltm': row['technocrat_trait_ltm'],
            'PresidentOfExecutive': row['PresidentOfExecutive']
        })
        current = stop + pd.Timedelta(days=1)
    return intervals
```

```python
primesmodel.loc[:,['PCM', 'start', 'end',
       'spell_duration', 'status', 'technocrat_trait_ltm','regimeFujimori','PresidentOfExecutive']].info()
```

```python
# Ensure 'start' and 'end' are datetime

primesmodel['id'] = primesmodel.PCM  # if not already assigned

# Generate the quarterly split spells
split_data = []
for _, row in primesmodel.iterrows():
    split_data.extend(split_spell_row(row))

# Create the final Andersen–Gill-style DataFrame
ag_primesmodel = pd.DataFrame(split_data)
ag_primesmodel['duration'] = (ag_primesmodel['stop'] - ag_primesmodel['start']).dt.days
ag_primesmodel['Quarter'] = ag_primesmodel['stop'].dt.to_period('Q').astype(str) 
ag_primesmodel
```
