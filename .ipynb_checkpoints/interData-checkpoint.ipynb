```python
# coding: utf-8

import pandas as pd
from urllib.parse import quote

# Pull tables from Wikipedia
base  = "https://es.wikipedia.org/wiki/" 
path  = "Anexo:Ministros_del_Interior_del_Perú"
url   = base + quote(path)          # Per%C3%BA

# Add a User-Agent header to mimic a browser and avoid 403 Forbidden error
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
intTables = pd.read_html(url, flavor="bs4",
                      attrs={"class": "wikitable sortable"}, storage_options={"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
```

# concat

```python
inta=pd.concat(intTables[:2],ignore_index=True)
```

```python
intb=pd.concat(intTables[2:4],ignore_index=True)
```

```python
intc=intTables[4]
inta=inta[~inta.isnull().all(axis=1)]
intb=intb[~intb.isnull().all(axis=1)]
intc=intc[~intc.isnull().all(axis=1)]
```

```python
inta.columns
```

```python
intb.columns
```

```python
intc.columns
```

```python
inta.columns=[b for a,b in inta.columns]
intb.columns=[b for a,b in intb.columns]
intc.columns=[b for a,b in intc.columns]
```

```python
intb["Fin"] = intb["Inicio"].shift(-1)
intb
```

```python
intc[["Inicio","Fin"]]=intc.Periodo.str.split("-",expand=True)
```

```python
inta.columns, intb.columns, intc.columns
```

```python
intc.drop(columns=["Periodo"],inplace=True)
intc.rename(columns={"Ministro":"Nombre"},inplace=True)
```

```python
inter=pd.concat([inta, intb, intc],ignore_index=True)
```

```python
inter.drop(columns=["Imagen"],inplace=True)
```

```python
import numpy as np
from datetime import datetime, timedelta
import locale

# 1. Configurar locale (con fallback por si estás en Windows)
try:
    locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")
except locale.Error:
    locale.setlocale(locale.LC_TIME, "Spanish_Spain.1252") # Windows

# Tu función lambda (agregamos validación por si hay NaN)
str_to_date = lambda x: datetime.strptime(x, "%d de %B de %Y") if pd.notna(x) else pd.NaT

inter[["Inicio", "Fin"]] = inter[["Inicio", "Fin"]].apply(lambda x: x.str.strip())
```

# Date format

```python
# 2. Convertir a datetime
inter['fecha_start'] = inter['Inicio'].apply(str_to_date)
inter['fecha_end'] = inter['Fin'].apply(str_to_date)



```
