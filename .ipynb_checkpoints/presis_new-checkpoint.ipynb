```python
# coding: utf-8

import pandas as pd
from urllib.parse import quote

# Pull tables from Wikipedia
base  = "https://es.wikipedia.org/wiki/"
path  = "Anexo:Gobernantes_del_Perú"
url   = base + quote(path)          # Per%C3%BA

# Add a User-Agent header to mimic a browser and avoid 403 Forbidden error
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
presisTables = pd.read_html(url, flavor="bs4", storage_options={"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
```

```python
presisAll=pd.concat([presisTables[0],presisTables[1],
           presisTables[5].drop(columns="Gobernante.1").rename(columns={"Gobernante.2":"Gobernante.1"})],
          ignore_index=True)
```

```python
presisAll.columns
```

```python
presisAll.drop(columns=["Gobernante"],inplace=True)
presisAll.rename(columns={"Gobernante.1":"Gobernante"}, inplace=True)
presisAll.head()
```

```python
presisAll.columns
```

```python
presisAll.columns=['Gobernante', 'Periodo', 'Vez', 'Mecanismo', 'Titulo','Organizacion']
```

```python
presisAll[presisAll.Periodo.str.contains(r'\[|\(',regex=True)]["Periodo"]
```

```python
regex_brackets = r'\[.*?\]|\(.*?\)'

presisAll['Periodo'] = presisAll['Periodo'].str.replace(regex_brackets, '', regex=True).str.strip()
```

```python
presisAll[presisAll.Periodo.str.split('-',expand=False).str.len()!=2]
```

```python
presisAll[['InicioGob',"FinGob"]]=presisAll.Periodo.str.split('-',expand=True)
```

```python
presisAll.head()
```

```python
presisAll[presisAll.InicioGob.str.split(r'\s*de\s*', expand=False).str.len()!=3]
```

```python
presisAll.loc[63,"InicioGob"]="13 de febrero de 1854"
```

```python
presisAll[presisAll.FinGob.str.split(r'\s*de\s*', expand=False).str.len()!=3]
```

```python
presisAll.loc[83,"FinGob"]="22 de julio de 1872"
presisAll.loc[109,"FinGob"]="18 de abril de 1904"
presisAll.loc[160,"FinGob"]="28 de julio de 2026"
```

```python
presisAll.loc[86,:]
```

```python
presisAll[['InicioGob', 'FinGob']] = presisAll[['InicioGob', 'FinGob']].apply(lambda x: x.str.strip())
```

```python
from datetime import datetime
import locale

# Set Spanish locale (use es_ES.utf8 or es_ES.UTF-8 depending on OS) - windows: "Spanish_Spain" or "Spanish_Spain.1252"
locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")

spanish_cellToGoodDates=lambda x: datetime.strptime(x, "%d de %B de %Y")

# Apply parsing with datetime and locale
presisAll[['startGob','endGob']]=presisAll[['InicioGob','FinGob']].map(spanish_cellToGoodDates)
presisAll
```

```python
presisAll.to_pickle("presisAll.pkl")
```
