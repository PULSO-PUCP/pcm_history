```python
import pandas as pd

pcm=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRzFOfy5dxsdPtHL5XyG9dDIegvpyD7ONvk5S9TplTIThM_WbJEATd7qxPKn7lioAK_kJ3sNQuVSNrO/pub?gid=970878581&single=true&output=csv")
```

```python
pcm["Titular"]=pcm.Titular.str.split("[",expand=True)[0]
```

```python
import numpy as np
from datetime import datetime, timedelta
import locale

# 1. Configurar locale (con fallback por si estás en Windows)
try:
    locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")
except locale.Error:
    locale.setlocale(locale.LC_TIME, "Spanish_Spain.1252") # Windows

# Tu función lambda (agregamos validación por si hay NaN)
str_to_date = lambda x: datetime.strptime(x, "%d de %B de %Y") if pd.notna(x) else pd.NaT

pcm[["Inicio", "Fin"]] = pcm[["Inicio", "Fin"]].apply(lambda x: x.str.strip())
```

```python
pcm['start'] = pcm['Inicio'].apply(str_to_date)
```

```python
pcm['end'] = pcm['Fin'].apply(str_to_date)
```

```python
pcm.columns
```

```python
keep=['Titular', 'Partido', 'Gobierno', 'Militar', 'regime',  'Categoría','start', 'end']
```

```python
pcm=pcm[keep]
```
