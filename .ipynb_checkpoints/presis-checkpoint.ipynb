```python
import pandas as pd
import re

# 1. Extract the year from FinGob (4 digits)
years = presisAll["FinGob"].str.extract(r"(\d{4})")[0]

# 2. Extract the month from FinGob (e.g., 'marzo', 'enero')
months = presisAll["FinGob"].str.extract(r"de\s+([a-zA-ZñÑ]+)\s+de")[0]


# 3. Smart fix function
def fix_col1(row, year, month):
    val = str(row["InicioGob"]).strip()

    # Case C: If it ALREADY ends with a 4-digit year, do nothing
    if re.search(r"\d{4}$", val):
        return val

    # Case A: If it's just a number (e.g., "10"), add month and year
    if val.isdigit():
        return f"{val} de {month} de {year}"

    # Case B: If it has text (e.g., "19 de enero") but no year, add year
    else:
        return f"{val} de {year}"


# 4. Apply and ASSIGN it back to the column
presisAll["InicioGob"] = presisAll.apply(
    lambda r: fix_col1(r, years[r.name], months[r.name]), axis=1
)
```

```python
from datetime import datetime
import locale
import pandas as pd

# Set Spanish locale
locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")

# Updated lambda function: Checks if x is valid before attempting to parse
spanish_cellToGoodDates = lambda x: (
    datetime.strptime(x, "%d de %B de %Y")
    if pd.notna(x) and isinstance(x, str)
    else None
)

# Apply parsing safely across the columns
presisAll[["startGob", "endGob"]] = presisAll[["InicioGob", "FinGob"]].map(
    spanish_cellToGoodDates
)

presisAll
```
