```python
import pandas as pd

gdps=pd.read_excel("Trimestral-20250518-151243.xlsx")
gdps
```

```python
gdps.columns=['trimestre','gdp']
gdps=gdps.iloc[1:,:]
gdps.reset_index(drop=True,inplace=True)
gdps
```

```python
def convert_t_quarter(quarter_str):
    """Converts 'T' followed by 3 digits (QYY) to 'YYYY-Qq' format."""
    if isinstance(quarter_str, str) and quarter_str.startswith('T') and len(quarter_str) == 4 and quarter_str[1:].isdigit():
        quarter = int(quarter_str[1])
        year_short = int(quarter_str[2:])
        # Determine the full year (assuming a reasonable range, e.g., 1900-2099)
        if 0 <= year_short <= 25:  # Adjust the upper bound as needed for your data
            year_full = 2000 + year_short
        else:
            year_full = 1900 + year_short
        return f"{year_full}Q{quarter}"
    else:
        return pd.NA  # Or handle invalid formats as needed

# Apply the conversion function to your 'quarter_code' column
gdps['pandas_quarter'] = gdps['trimestre'].apply(convert_t_quarter)

gdps[['trimestre', 'pandas_quarter','gdp']].head()
```

```python
gdps.to_excel('gdps.xlsx', index=True)
gdps.to_parquet('gdps.parquet')
```
