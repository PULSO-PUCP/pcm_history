# Date format

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import locale

# 1. Configurar locale (con fallback por si estás en Windows)
try:
    locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")
except locale.Error:
    locale.setlocale(locale.LC_TIME, "Spanish_Spain.1252") # Windows

# Tu función lambda (agregamos validación por si hay NaN)
str_to_date = lambda x: datetime.strptime(x, "%d de %B de %Y") if pd.notna(x) else pd.NaT

# --- Supongamos que tu columna de fechas se llama 'fecha_inicio' ---

# 2. Convertir a datetime
mefa['fecha_start'] = mefa['Fecha de inicio'].apply(str_to_date)

# 3. Desplazar (-1), restar 1 día y volver a formato texto español
mefa['fecha_end'] = (mefa['fecha_start'].shift(-1) - timedelta(days=1))



mefa
```
